```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset Load
df = pd.read_csv("Accident_Information.csv")

# Basic Information
print("Rows and Columns:", df.shape)
print(df.head())

# Numerical Columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Graph 1: First Numerical Column Distribution
plt.figure(figsize=(8,5))
df[num_cols[0]].value_counts().head(10).plot(kind='bar')
plt.title(f'{num_cols[0]} Distribution')
plt.savefig('graph1.png')
plt.show()

# Graph 2: Second Numerical Column Distribution
if len(num_cols) > 1:
    plt.figure(figsize=(8,5))
    df[num_cols[1]].value_counts().head(10).plot(kind='bar')
    plt.title(f'{num_cols[1]} Distribution')
    plt.savefig('graph2.png')
    plt.show()

# Graph 3: Correlation Heatmap
plt.figure(figsize=(8,6))
corr = df[num_cols].corr()
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.title('Correlation Heatmap')
plt.savefig('graph3.png')
plt.show()
```
